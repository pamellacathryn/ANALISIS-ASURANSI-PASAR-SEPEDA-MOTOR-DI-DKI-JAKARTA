# BAGIAN 1: Simulasi Bulanan Transaksi Perusahaan Asuransi dengan Konsumen

```python
#Fungsi Permintaan
#Import Library
import numpy as np
import matplotlib.pyplot as plt

p_1=np.arange(0,101,1) #harga (dalam juta)
q_1=[0 for i in range(len(p_1))] #kuantitas

for i in range(len(p_1)):
  q_1[i]=4100099.497/(p_1[i]*(10**6))

fig, ax=plt.subplots(figsize=(12,7.5))
ax.plot(q_1,p_1, label="qk(p)")
ax.set_title("Fungsi Permintaan Sepeda Motor")
ax.set_xlabel("Kuantitas")
ax.set_ylabel("Harga Sepeda Motor (juta rupiah)")
ax.legend()
ax.grid(True)
ax.set_ylim(0,21)
plt.show()
```

```python
#Simulasi Permintaan
import numpy as np
np.random.seed(4)
p = 0.0005
bulan = 12
damage = 3047.818
premi = damage*100*p*12
r = p*100
berhasil = 0
gagal = 0
total_profit_loss = 0
for i in range (10):
    print("Simulasi tahun ke-", str(i+1))
    claim = np.random.poisson(r, bulan)
    print(claim)
    claim = sum(claim)
    insured = claim*3047.818
    profit_loss = premi - insured
    total_profit_loss = total_profit_loss + profit_loss
    print("%12s %12s %12s %12s" % ("Total Premium", "Insured", "Profit/Loss", "Claim"))
    print("%12.1f %12.1f %12.1f %12.1f" % (premi, insured, profit_loss, claim))
    if (premi > insured ):
        print("Asuransi berhasil bayar")
        berhasil += 1
    else:
        print("Asuransi gagal bayar")
        gagal += 1
    print("")
print("Asuransi berhasil bayar sebanyak ", str(berhasil), " kali dan asuransi gagal bayar sebanyak ", str(gagal), " kali")
print("Selama simulasi 10 tahun, asuransi rugi/untung sebanyak ", str(total_profit_loss))
print("")
count = 0
for i in range (100000):
    claim = np.random.poisson(r, bulan)
    total_claim = sum(claim)
    if total_claim*3000 > premi:
        count += 1
p = count/100000
print("Hasil simulasi selama 100.000 tahun, peluang gagal bayar adalah ", p)
```

# BAGIAN 2: Simulasi Bulanan Transaksi Perusahaan Asuransi dengan Produsen




```python
np.random.seed(3)
#Simulasi Penawaran
import numpy as np
p = 2/9
bulan = 12
damage = 1000000000
premi = damage*100*p*12
r = p
mu = 100*r
sigma = 4.157
berhasil = 0
gagal = 0
total_profit_loss = 0
for i in range (10):
    print("Simulasi tahun ke-", str(i+1))
    claim = np.random.normal(mu, sigma, bulan)
    print(claim)
    claim = sum(claim)
    claim = int(claim)
    insured = claim * damage
    profit_loss = premi - insured
    total_profit_loss = total_profit_loss + profit_loss
    print("%12s %12s %12s %12s" % ("Total Premium", "Insured", "Profit/Loss", "Claim"))
    print("%12.1f %12.1f %12.1f %12.1f" % (premi, insured, profit_loss, claim))
    if (premi > insured ):
        print("Asuransi berhasil bayar")
        berhasil += 1
    else:
        print("Asuransi gagal bayar")
        gagal += 1
    print("")
print("Asuransi berhasil bayar sebanyak ", str(berhasil), " kali dan asuransi gagal bayar sebanyak ", str(gagal), " kali")
print("Selama simulasi 10 tahun, asuransi rugi/untung sebanyak ", str(total_profit_loss))
print("")
print("Hasil simulasi selama 10 tahun, peluang gagal bayar adalah ", gagal/(gagal+berhasil))
```

# BAGIAN 3: Kurva Permintaan, Kurva Penawaran, dan Apabila Kegiatan Konsumen dan Produsen Terganggu

```python
#Grafik kurva permintaan
#Import Library
import numpy as np
import matplotlib.pyplot as plt

p_1=np.arange(0,101,1) #harga (dalam juta)
q_1=[0 for i in range(len(p_1))] #kuantitas

for i in range(len(p_1)):
  q_1[i]=1303831640/(p_1[i]*(10**6))

fig, ax=plt.subplots(figsize=(12,7.5))
ax.plot(q_1,p_1, label="qk(p)")
ax.set_title("Fungsi Permintaan Pasar Konsumen")
ax.set_xlabel("Kuantitas")
ax.set_ylabel("Harga Sepeda Motor (juta rupiah)")
ax.legend()
ax.grid(True)
ax.set_ylim(0,21)
plt.show()
```

```python
#Grafik Penawaran
import numpy as np
import matplotlib.pyplot as plt

p_1=np.arange(0,101,1) #harga (dalam juta)
q_1=[0 for i in range(len(p_1))] #kuantitas

for i in range(len(p_1)):
  q_1[i]=1.9*10**-3*(p_1[i]*(10**6))**(2/3)

fig, ax=plt.subplots(figsize=(12,7.5))
ax.plot(q_1,p_1, label="qk(p)")
ax.set_title("Fungsi Penawaran Sepeda Motor")
ax.set_xlabel("Kuantitas")
ax.set_ylabel("Harga Sepeda Motor (juta rupiah)")
ax.legend()
ax.grid(True)
ax.set_ylim(0,21)
plt.show()
```

```python
#Mencari Titik Equilibrium dan Menggambarkan Grafiknya
import numpy as np
import matplotlib.pyplot as plt

Price=np.arange(1,20000) #harga dalam 10^3
Q_supply_arr=[]
Q_demand_arr=[]
for P in Price:
    Q_supply = (1.9)*(10**(-3))*((P*(10**3))**(2/3))
    Q_demand = 1303831640/(P*(10**3))
    if abs(Q_supply - Q_demand) <= 0.01:
        print("Harga : ", "Rp"+str(P)+"000,00")
        print("Q_supply : ", Q_supply)
        print("Q_demand : ", Q_demand)
    Q_supply_arr.append(Q_supply)
    Q_demand_arr.append(Q_demand)

fig, ax=plt.subplots()
plt.plot(Q_supply_arr, Price)
plt.plot(Q_demand_arr, Price)
ax.set_title("Fungsi Penawaran dan Permintaan Sepeda Motor")
ax.set_xlabel("Kuantitas")
ax.set_ylabel("Harga Sepeda Motor (10^3)")
ax.legend(["Supply", "Demand"])
ax.grid(True)
ax.set_xlim(0,300)
ax.set_ylim(0,20000)
ax.xaxis.set_ticks(np.arange(0, 300, 25))
ax.yaxis.set_ticks(np.arange(0, 20000, 1250))

plt.show()
```

```python
#Kegiatan Konsumsi Terganggu
#Grafik kurva permintaan terganggu

#Import Library
import numpy as np
import matplotlib.pyplot as plt

Price=np.arange(1,20000) #harga dalam 10^3
Q_supply_arr=[]
Q_new_demand_arr=[]
Q_demand_arr=[]
for P in Price:
    Q_supply = (1.9)*(10**(-3))*((P*(10**3))**(2/3))
    Q_new_demand = 1241763416/(P*(10**3))
    Q_demand = 1303831640/(P*(10**3))
    if abs(Q_supply - Q_new_demand) <= 0.01:
        print("Harga : ", "Rp"+str(P)+"000,00")
        print("Kuantitas : ", Q_new_demand)
    Q_supply_arr.append(Q_supply)
    Q_new_demand_arr.append(Q_new_demand)
    Q_demand_arr.append(Q_demand)

fig, ax=plt.subplots()
plt.plot(Q_supply_arr, Price)
plt.plot(Q_new_demand_arr, Price)
plt.plot(Q_demand_arr, Price, linestyle = 'dashed')
ax.set_title("Fungsi Penawaran-Permintaan Sepeda Motor Jika Kegiatan Konsumsi Terganggu")
ax.set_xlabel("Kuantitas")
ax.set_ylabel("Harga Sepeda Motor (10^3)")
ax.legend(["Supply", "New Demand", "Demand"])
ax.grid(True)
ax.set_xlim(0,300)
ax.set_ylim(0,20000)
ax.xaxis.set_ticks(np.arange(0, 300, 25))
ax.yaxis.set_ticks(np.arange(0, 20000, 1250))

plt.show()
```

```python
#Kegiatan Produksi Terganggu
#Grafik kurva penawaran terganggu
#Import Library
import numpy as np
import matplotlib.pyplot as plt

Price=np.arange(1,20000) #harga dalam 10^3
Q_supply_arr=[]
Q_new_supply_arr=[]
Q_demand_arr=[]
for P in Price:
    Q_supply = (1.9)*(10**(-3))*((P*(10**3))**(2/3))
    Q_new_supply = (6.383*(10**(-4)))*((P*(10**3))**(41/59))
    Q_demand = 1303831640/(P*(10**3))
    if abs(Q_new_supply - Q_demand) <= 0.005:
        print("Harga : ", "Rp"+str(P)+"000,00")
        print("Kuantitas : ", Q_new_supply)
    Q_supply_arr.append(Q_supply)
    Q_new_supply_arr.append(Q_new_supply)
    Q_demand_arr.append(Q_demand)

fig, ax=plt.subplots()
plt.plot(Q_supply_arr, Price, linestyle = 'dashed')
plt.plot(Q_new_supply_arr, Price)
plt.plot(Q_demand_arr, Price)
ax.set_title("Fungsi Penawaran-Permintaan Sepeda Motor Jika Kegiatan Produksi Terganggu")
ax.set_xlabel("Kuantitas")
ax.set_ylabel("Harga Sepeda Motor (10^3)")
ax.legend(["Supply", "New Supply", "Demand"])
ax.grid(True)
ax.set_xlim(0,300)
ax.set_ylim(0,20000)
ax.xaxis.set_ticks(np.arange(0, 300, 25))
ax.yaxis.set_ticks(np.arange(0, 20000, 1250))

plt.show()
```
